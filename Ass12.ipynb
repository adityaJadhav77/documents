```python
import random

class Agent : 
  def __init__(self , name) :
    self.name = name
  
  def act(self,state) : 
    actions = state.get("actions" , ["cooperate" , "defect"])
    return random.choice(actions)
  
def simulate(steps=10) :
  state = {"actions" : ["cooperate" , "defect"] , "score" : {"A" : 0 , "B" : 0}}
  A = Agent("A")
  B = Agent("B")

  history = []
  for t in range(steps) : 
    a1 = A.act(state)
    a2 = B.act(state)

    if a1 == "cooperate" and a2 == "cooperate" : 
      state["score"]["A"] += 3 ; 
      state["score"]["B"] += 3
    elif a1 == "cooperate" and a2 == "defect" : 
      state["score"]["A"] += 0 ; 
      state["score"]["B"] += 5
    elif a1 == "defect" and a2 == "cooperate" : 
      state["score"]["A"] += 5 ; 
      state["score"]["B"] += 0
    else:
      state["score"]["A"] += 1 ; 
      state["score"]["B"] += 1

    history.append((t , a1 , a2 , dict(state["score"])))

  return history

def main() : 
  hist = simulate(steps=20)
  for h in hist : 
    print(h) 

if __name__ == "__main__" : 
  main()    
```
